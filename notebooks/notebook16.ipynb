```python
R = np.full((256, 256), 255)
G = np.zeros((256, 256))
B = np.zeros((256, 256))

rgb = np.dstack((R, G, B))

plt.imshow(rgb)
plt.title("Pure red")
plt.show()
```

```python
x = np.arange(256).reshape(1,256)

R = x  #+ np.zeros((256,1))
G = np.zeros((256,256))
B = 255 - x  #+ np.zeros((256,1))

rgb = np.dstack((R,G,B)).astype(np.uint8)

plt.imshow(rgb)
plt.title("Red to blue gradient")
plt.show()
```

```python
# IGNORE, JUST FOR FUN
x = np.arange(256).reshape(1,256)
y = np.arange(256).reshape(256,1)

# do something random with the x and y values and assign them to Red, Green, and Blue
R = (x + y) % 256
G = (2 * x + 0 * y) % 256
B = (2 * y + 0 * x) % 256

# put them together as pixels
rgb = np.dstack((R,G,B)).astype(np.uint8)

plt.imshow(rgb)
plt.title("Generated color pattern")
plt.show()
```

```python
x = np.arange(256).reshape(1,256)
y = np.arange(256).reshape(256,1)

R = x + 0*y
G = y + 0*x
B = np.zeros((256,256))

fig, ax = plt.subplots(1,3, figsize=(10,3))

ax[0].imshow(R, cmap="gray", vmin=0, vmax=255)
ax[0].set_title("Red channel")

ax[1].imshow(G, cmap="gray", vmin=0, vmax=255)
ax[1].set_title("Green channel")

ax[2].imshow(B, cmap="gray", vmin=0, vmax=255)
ax[2].set_title("Blue channel")

plt.show()
```

```python
rgb = np.dstack((R,G,B)).astype(np.uint8)

plt.imshow(rgb)
plt.title("Color image from three arrays")
plt.show()
```
