```python
tea = np.array([21,10,22])
coffee = np.array([32,38,13])
soda = np.array([49,50,28])
juice = np.array([34,13,23])

# try to put them together as in previous array
stacked_preferences = np.stack((tea,coffee,soda,juice))
print(stacked_preferences)
print(stacked_preferences.shape)
```

```python
# PARANTHETICALLY
# .T -- TRANSPOSE swaps rows and columns
stacked_preferences_transposed = stacked_preferences.T
print(stacked_preferences_transposed)
print(stacked_preferences_transposed.shape)
```

```python
# ANOTHER TRANSPOSE EXAMPLE
a = np.arange(6).reshape(2,3)

print(a)
print(a.sum(axis=0))
print(a.T)
print(a.T.sum(axis=1))
# NOW BACK TO REGULARLY SCHEDULED PROGRAMMING
```

```python
# before the transpose break, what we had didn't quite work. it added a new ROW dimension, put it together along the wrong axis
stacked_preferences = np.stack((tea,coffee,soda,juice),axis=1)    # axis = 1 does a column stack
print(stacked_preferences)
```

```python
# equivalent to column_stack
stacked_preferences = np.column_stack((tea,coffee,soda,juice))
print(stacked_preferences)
```

```python
# what if different data types? NumPy will upcast everything to a common data type
a = np.array([1, 2, 3])          # ints
b = np.array(["A", "B", "C"])    # strings

c = np.stack((a, b), axis=1)
print(c)
```

```python
# great if just printing multiple arrays together side by side
print(np.stack((a,b),axis=1))
```

# NumPy uses shape; pandas uses labels

```python
# NOW LET'S DO IT THE PANDAS WAY
participants_per_study_in_df = pref_df.sum(axis=1)  
# numpy is values without labels, so we needed keepdims to tell us if it was a row or column
# pandas returns a Series which includes a label, which tells it what's what
print(participants_per_study_in_df)
```

```python
percent_by_row_df = pref_df / participants_per_study_in_df
percent_by_row_df *= 100
print(percent_by_row_df.round(1))
```

# The Series labels are row labels
# DataFrame expects column labels for alignment
so it fails

```python
print(type(percent_by_row_df))
#print(percent_by_row_df.shape)
print(type(participants_per_study_in_df))
#print(participants_per_study_in_df.shape)
```

```python
# TELL PANDAS TO DO IT ROW-WISE.
# axis = 0: match by row labels
percent_by_row_df = pref_df.div(participants_per_study_in_df, axis=0)
percent_by_row_df *= 100
print(percent_by_row_df.round(1))
```

```python
# A SECOND LOOK FOR COMPARISON
pref = np.array([[21, 32, 49, 34],
              [10, 38, 50, 13],
              [22, 13, 28, 23]])

# ONE WAY TO CREATE A DATAFRAME, SEND IT A NUMPY ARRAY, A LIST OF COLUMN NAMES, A LIST OF INDEX (ROW) NAMES
pref_df = pd.DataFrame(
    pref,
    columns=["Tea","Coffee","Soda","Juice"],
    index=["Study A","Study B","Study C"]
)
```

```python
# ANOTHER WAY TO CREATE A DATAFRAME: SEND THE CONSTRUCTOR A DICTIONARY
# pandas DataFrames are heterogeneous across columns
df = pd.DataFrame({
    "Name": ["A", "B", "C"],
    "AQI": [72, 85, 60]
})

print(df.dtypes)
```

# ADDITIONAL CONSTRUCTOR VARIATIONS

```python
# similar to
dict1 = {}
dict1["Name"] = ["A", "B", "C"]
dict1["AQI"] = [72, 85, 60]

# can send that dictionay to constructor
df = pd.DataFrame(dict1)
```

```python
# create DataFrame out of rows:
df = pd.DataFrame([
    {"Name": "A", "AQI": 72},
    {"Name": "B", "AQI": 85},
    {"Name": "C", "AQI": 60}
])
```

```python
# you can include row names, too, by adding index parameter
df = pd.DataFrame(
    {
        "Name": ["A", "B", "C"],
        "AQI": [72, 85, 60]
    },
    index=["Study A", "Study B", "Study C"]
)

print(df)
```

```python
# OR set it separately
df = pd.DataFrame(
    {
        "Name": ["A", "B", "C"],
        "AQI": [72, 85, 60]
    })
df.index = ["Study A", "Study B", "Study C"]
print(df)
```

```python
# OR EVEN
pd.DataFrame(
    {
        "Study A": {"Name": "A", "AQI": 72},
        "Study B": {"Name": "B", "AQI": 85}
    }
).T
```

```python
# can i specify column names?
df = pd.DataFrame(
    {
        ["A", "B", "C"],
        [72, 85, 60]
    },
    columns=["Name","AQI"]
)

print(df)
```

```python
# the above is NOT VALID.
# if you pass a dictionary, keys = column names
# if you want to use columns, data must NOT be a dict
# CAN use a list of lists (OR NUMPY ARRAY as in our first example)

df = pd.DataFrame(
    [
        ["A", 72],
        ["B", 85],
        ["C", 60]
    ],
    columns=["Name", "AQI"],
    index=["Study A", "Study B", "Study C"]
)
print(df)
```
