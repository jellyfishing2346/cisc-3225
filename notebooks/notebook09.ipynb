```python
import numpy as np

rng = np.random.default_rng()
grades = rng.normal(85,7,23).astype(np.int64)
print(grades)
```

```python
grades[0] += 10
print(grades)

grades[0] -= 10 # reset zeroth student
print(grades)
```

```python
# how to add 10 to all students' grades?
```

```python
grades[[0,9,12]] += 10
print(grades)
```

```python
grades[[3,19]] += 5
print(grades)
```

```python
final_exam = grades2d[:, 2]  # third column (exam 3)
final_exam
```

```python
pass_mask = final_exam >= 80
print("pass_mask:", pass_mask)
```

```python
print(f"length of final_exam array: {final_exam.size}")
print(f"length of pass_mask array: {pass_mask.size}")
```

```python
print("passing scores:", final_exam[pass_mask])
print("count passing:", final_exam[pass_mask].size)
```

```python
# Scores in a "B range": 80–89 inclusive (1D example)
mask_B = (final_exam >= 80) & (final_exam <= 89)
print("B-range scores:", final_exam[mask_B])
```

```python
# Low OR High (1D example)
mask_low_or_high = (final_exam < 70) | (final_exam >= 90)
print("low OR high:", final_exam[mask_low_or_high])
```

```python
print("final_exam min:", final_exam.min())
print("final_exam max:", final_exam.max())
print("final_exam mean:", final_exam.mean())
print("final_exam sum:", final_exam.sum())
```
