```python
import pandas as pd
import numpy as np

df = pd.read_csv("presidential_heights.csv")    # df is a pandas DATAFRAME

names = df["last_name"].to_numpy()
heights = df["height_inches"].to_numpy()

order = np.argsort(heights)
print(names[order])
print(heights[order])

# what does order[-1] represent

for n, h in zip(names[order],heights[order]):
    print(f"{n}: {h}")
```

```python
df2 = pd.read_csv("presidential_heights_better.csv")    # df is a pandas DATAFRAME

numbers = df2["number"].to_numpy()
names = df2["last_name"].to_numpy()
heights = df2["height_inches"].to_numpy()

order = np.argsort(heights)

for num,name,ht in zip(numbers[order],names[order],heights[order]):
    print(f"{num} {name} {ht}")
```

```python
import matplotlib.pyplot as plt

plt.bar(names, heights)
plt.xlabel("President")
plt.ylabel("Height (inches)")
plt.title("Presidential Heights")
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.bar(names,heights)
plt.xticks(rotation=90)
plt.ylabel("Height (inches)")
plt.title("Presidential Heights")
plt.show()
```

```python
order = np.argsort(heights)

sorted_names = names[order]
sorted_heights = heights[order]
```

```python
plt.figure(figsize=(10,4))
plt.bar(sorted_names,sorted_heights)
plt.xticks(rotation=90)
plt.ylabel("Height (inches)")
plt.title("Presidential Heights (Sorted)")
plt.show()
```

```python
labels = [f"{num} {name}" for num, name in zip(numbers, names)]

plt.figure(figsize=(12, 4))
plt.bar(labels, heights)
plt.xticks(rotation=90)
plt.ylabel("Height (inches)")
plt.title("Presidential Heights")
plt.show()
```

```python
order = np.argsort(heights)

sorted_heights = heights[order]
sorted_names = names[order]
sorted_numbers = numbers[order]
labels = [f"{num} {name}" for num, name in zip(sorted_numbers, sorted_names)]

plt.figure(figsize=(12,4))
plt.bar(labels,sorted_heights)
plt.xticks(rotation=90)
plt.ylabel("Height (inches)")
plt.title("Presidential Heights (Sorted)")
plt.show()
```

```python
order = np.argsort(heights)[::-1]

sorted_heights = heights[order]
sorted_names = names[order]
sorted_numbers = numbers[order]
labels = [f"{num} {name}" for num, name in zip(sorted_numbers, sorted_names)]

plt.figure(figsize=(12,4))
plt.bar(labels,sorted_heights)
plt.xticks(rotation=90)
plt.ylabel("Height (inches)")
plt.title("Presidential Heights (Sorted)")
plt.show()
```
