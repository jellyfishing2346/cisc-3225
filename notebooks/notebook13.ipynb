```python
# row normalization / centering
A = np.array([[1,2,3],
              [4,5,6]])

row_means = np.mean(A, axis=1)

print(A.shape)
print(row_means.shape)
print(A - row_means)
```

```python
A = np.array([[1,2,3],
              [4,5,6]])

row_means = np.mean(A, axis=1, keepdims=True)

print(A.shape)
print(row_means.shape)
print(A - row_means)
```
