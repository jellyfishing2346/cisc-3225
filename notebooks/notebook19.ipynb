```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Study": ["A","A","B","B","C","C"],
    "Drink": ["Coffee","Tea","Coffee","Tea","Coffee","Tea"],
    "Amount": [30,10,25,20,35,15]
})
```

```python
df
```

```python
df[df["Drink"] == "Coffee"]
```

```python
coffee = df[df["Drink"] == "Coffee"].groupby("Study")["Amount"].sum()
```

```python
studytotal = df.groupby("Study")["Amount"].sum()
```

```python
coffee / studytotal * 100
```
